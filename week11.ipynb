```python
#실습 문제 7

command = input("명령어 입력: ")
com = command.split()
if com[0] == "wc":    
    dic = {}
    LCount = 0;
    f = open(com[1],'r')
    lines = f.readlines()
    for line in lines:
        LCount+=1;
        words = line.split()
        for word in words:
            if word in dic:
                dic[word]+=1;
            else:
                dic[word] = 1;
    f.close()
        

print("라인 수:", LCount)
for word in dic:
    print(word, dic[word])
```

```python
#실습 문제 8
# -*- coding: cp949 -*-
class Calc:
    def sum(self,a,b):
        result=a+b
        print("{0:d} + {1:d} = {2:d} 입니다.".format(a,b,result))
    def sub(self,a,b):
        result=a-b
        print("{0:d} - {1:d} = {2:d} 입니다.".format(a,b,result))
    def mul(self,a,b):
        result=a*b
        print("{0:d} * {1:d} = {2:f} 입니다.".format(a,b,result))
    def div(self,a,b):
        result=a/b
        print("{0:d} / {1:d} = {2:f} 입니다.".format(a,b,result))

calc = Calc()
calc.sum(1,2)
calc.sub(8,6)
calc.mul(4,2)
calc.div(6,3)
```

```python
#실습 문제 9

class Stack:
    def __init__ (self):
        self.top = []
    
    def push (self, item):
        self.top.append(item)

    def pop(self):
        if not self.isEmpty():
            return self.top.pop(-1)
        else:
            print("Stack underflow")
            exit()

    def clear(self):
        self.top=[]
            
    def isEmpty(self):
        return len(self.top)==0

    def size(self):
        return len(self.top)
    
q = input("계산식을 입력하세요: ")

elem = q.split()
num = Stack()
op = Stack()

for i in elem:
    if i>='0' and i<='9':
        num.push(i)
    else:
        op.push(i)
        
number = [0 for i in range(num.size())]
operator = [0 for i in range(op.size())]

nsize = num.size()
osize = op.size()

for i in range(nsize):
    number[nsize-i-1] = num.pop()

for i in range(op.size()):
    operator[osize-i-1] = op.pop()
    
num.clear()
op.clear()

for i in range (len(number)):

    num.push(number[i])
    
    if i < len(operator):
        if num.size()>=2 and op.size()>=1 and operator[i] != '*' and operator[i] != '/':
            while not op.isEmpty():
                
                n1 = int(num.pop())
                n2 = int(num.pop())
                op1 = op.pop()
    
                if op1=="+":
                    num.push(n2+n1)
                elif op1=="-":
                    num.push(n2-n1)
                elif op1=="*":
                    num.push(n2*n1)
                elif op1=="/":
                    num.push(n2/n1)
    
        op.push(operator[i])
        
    else:
        if num.size()>=2 and op.size()>=1:
            while not op.isEmpty():
                n1 = int(num.pop())
                n2 = int(num.pop())
                op1 = op.pop()
    
                if op1=="+":
                    num.push(n2+n1)
                elif op1=="-":
                    num.push(n2-n1)
                elif op1=="*":
                    num.push(n2*n1)
                elif op1=="/":
                    num.push(n2/n1)

print(num.pop())
```
