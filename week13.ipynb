```python
#실습문제1

import csv
import matplotlib.pyplot as plt

f = open('seoul2.csv')
data = csv.reader(f)
next(data)

birthMax = []
birthMin = []

for row in data:
    month = row[0].split('-')[1]
    date = row[0].split('-')[2]
    if row[-1] != '':
        if month=='06' and date=='16':
            birthMax.append(float(row[-1]))
            birthMin.append(float(row[-2]))
            
plt.plot(birthMax, 'r', label='Max')
plt.plot(birthMin, 'b', label='Min')
plt.legend()
plt.show()
```

```python
#실습문제2

import csv
import matplotlib.pyplot as plt

f = open('seoul2.csv')
data = csv.reader(f)
next(data)

jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))

plt.hist(jan, bins=100, color='b', label='Jan')
plt.hist(aug, bins=100, color='r', label='Aug')
plt.legend()
plt.show()
```
